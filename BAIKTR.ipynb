```python
import numpy as np

# Xác định các quantile của cột 'truong_4'
q1 = df['truong_4'].quantile(0.25)
q3 = df['truong_4'].quantile(0.75)
iqr = q3 - q1

# Tính toán giới hạn dưới và giới hạn trên
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Capping các giá trị ngoại lệ
df['truong_4'] = np.where(df['truong_4'] < lower_bound, lower_bound, df['truong_4'])
df['truong_4'] = np.where(df['truong_4'] > upper_bound, upper_bound, df['truong_4'])
```

### Xây dựng mô hình

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tập tin CSV đã được tiền xử lý
df = pd.read_csv('data_processed.csv')

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
train_data, test_data = train_test_split(df['truong_4'], test_size=0.2, shuffle=False)

# Tạo và huấn luyện mô hình SARIMA với tập huấn luyện
model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
result = model.fit()

# Dự đoán trên tập kiểm tra
forecast = result.predict(start=len(train_data), end=len(train_data) + len(test_data) - 1, typ='levels')

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Tính toán sai số tuyệt đối trung bình (MAE)
mae = mean_absolute_error(test_data, forecast)

# Tính toán sai số bình phương trung bình (MSE)
mse = mean_squared_error(test_data, forecast)



# Hiển thị các kết quả
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
# Tính toán Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Hiển thị kết quả RMSE
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Import thư viện matplotlib
import matplotlib.pyplot as plt

# Đặt kích thước của biểu đồ
plt.figure(figsize=(10, 6))

# Vẽ dữ liệu thực tế
plt.plot(test_data.index, test_data, label='Dữ liệu thực tế', color='blue')

# Vẽ dự đoán
plt.plot(test_data.index, forecast, label='Dự đoán', color='red')

# Đặt tiêu đề và nhãn trục
plt.title('So sánh giữa dữ liệu thực tế và dự đoán')
plt.xlabel('Thời gian')
plt.ylabel('Giá trị')

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tập tin CSV đã được tiền xử lý
df = pd.read_csv('data_processed.csv')

# Chuyển cột 'date' sang định dạng datetime và đặt làm chỉ số
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
train_data, test_data = train_test_split(df, test_size=0.2, shuffle=False)

# Tạo và huấn luyện mô hình ARIMAX với tập huấn luyện
model = ARIMA(train_data['truong_4'], order=(15, 1, 15))  # Tăng bậc AR và MA
result = model.fit()

# Dự đoán trên tập kiểm tra
forecast = result.predict(start=test_data.index[0], end=test_data.index[-1], typ='levels')


# Tính toán sai số tuyệt đối trung bình (MAE)
mae = mean_absolute_error(test_data['truong_4'], forecast)

# Tính toán sai số bình phương trung bình (MSE)
mse = mean_squared_error(test_data['truong_4'], forecast)

# Tính toán Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Hiển thị các kết quả
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Vẽ biểu đồ so sánh giữa dữ liệu thực tế và dự đoán
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data['truong_4'], label='Dữ liệu thực tế')
plt.plot(test_data.index, forecast, label='Dự đoán', color='red')
plt.title('So sánh giữa dữ liệu thực tế và dự đoán')
plt.xlabel('Thời gian')
plt.ylabel('Giá trị')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Đọc dữ liệu từ file CSV
data = pd.read_csv('data_processed.csv')

# Chuyển đổi cột 'date' sang định dạng datetime và đặt làm index
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)


# Thực hiện phân tích phân loại thời gian
decomposition = seasonal_decompose(data['truong_4'], model='additive')

# Trực quan hóa các thành phần
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(decomposition.observed, label='Observed')
plt.legend()

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend()

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend()

plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```
