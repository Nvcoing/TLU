```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# Äá»c dá»¯ liá»‡u
df = pd.read_csv('bt.csv')

# Decompose the time series for each variable
decomposition_8 = seasonal_decompose(df['8'], period=12, model='additive')
fig_8 = decomposition_8.plot()
plt.title('Decomposition of Variable 8')
plt.show()

decomposition_9 = seasonal_decompose(df['9'], period=12, model='additive')
fig_9 = decomposition_9.plot()
plt.title('Decomposition of Variable 9')
plt.show()

decomposition_10 = seasonal_decompose(df['10'], period=12, model='additive')
fig_10 = decomposition_10.plot()
plt.title('Decomposition of Variable 10')
plt.show()

# Plot the autocorrelation and partial autocorrelation for each variable
fig, ax = plt.subplots(6, 1, figsize=(12, 24))

sm.graphics.tsa.plot_acf(df['8'], lags=50, ax=ax[0])
ax[0].set_title('Autocorrelation of Variable 8')

sm.graphics.tsa.plot_pacf(df['8'], lags=50, ax=ax[1])
ax[1].set_title('Partial Autocorrelation of Variable 8')

sm.graphics.tsa.plot_acf(df['9'], lags=50, ax=ax[2])
ax[2].set_title('Autocorrelation of Variable 9')

sm.graphics.tsa.plot_pacf(df['9'], lags=50, ax=ax[3])
ax[3].set_title('Partial Autocorrelation of Variable 9')

sm.graphics.tsa.plot_acf(df['10'], lags=50, ax=ax[4])
ax[4].set_title('Autocorrelation of Variable 10')

sm.graphics.tsa.plot_pacf(df['10'], lags=50, ax=ax[5])
ax[5].set_title('Partial Autocorrelation of Variable 10')

plt.tight_layout()
plt.show()

# Plot rolling statistics for each variable
rolling_mean_8 = df['8'].rolling(window=12).mean()
rolling_std_8 = df['8'].rolling(window=12).std()

rolling_mean_9 = df['9'].rolling(window=12).mean()
rolling_std_9 = df['9'].rolling(window=12).std()

rolling_mean_10 = df['10'].rolling(window=12).mean()
rolling_std_10 = df['10'].rolling(window=12).std()

plt.figure(figsize=(12, 8))
plt.plot(df['8'], label='Variable 8')
plt.plot(rolling_mean_8, label='Rolling Mean (Variable 8)')
plt.plot(rolling_std_8, label='Rolling Std (Variable 8)')

plt.plot(df['9'], label='Variable 9')
plt.plot(rolling_mean_9, label='Rolling Mean (Variable 9)')
plt.plot(rolling_std_9, label='Rolling Std (Variable 9)')

plt.plot(df['10'], label='Variable 10')
plt.plot(rolling_mean_10, label='Rolling Mean (Variable 10)')
plt.plot(rolling_std_10, label='Rolling Std (Variable 10)')

plt.title('Rolling Mean & Standard Deviation')
plt.legend()
plt.show()

# Check for stationarity using the Augmented Dickey-Fuller test for each variable
result_8 = adfuller(df['8'])
print('\nADF Statistic (Variable 8): %f' % result_8[0])
print('p-value (Variable 8): %f' % result_8[1])
print('Critical Values (Variable 8):')
for key, value in result_8[4].items():
    print('\t%s: %.3f' % (key, value))

result_9 = adfuller(df['9'])
print('\nADF Statistic (Variable 9): %f' % result_9[0])
print('p-value (Variable 9): %f' % result_9[1])
print('Critical Values (Variable 9):')
for key, value in result_9[4].items():
    print('\t%s: %.3f' % (key, value))

result_10 = adfuller(df['10'])
print('\nADF Statistic (Variable 10): %f' % result_10[0])
print('p-value (Variable 10): %f' % result_10[1])
print('Critical Values (Variable 10):')
for key, value in result_10[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
import pandas as pd
import numpy as np
from pykalman import KalmanFilter
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Äá»c dá»¯ liá»‡u
df = pd.read_csv('bt.csv')

# Láº¥y cÃ¡c cá»™t '8', '9', '10' vÃ  chuyá»ƒn thÃ nh máº£ng numpy
measurements = df[['8', '9', '10']].values

# Táº¡o mÃ´ hÃ¬nh Kalman filter
kf = KalmanFilter(initial_state_mean=np.zeros(3), n_dim_obs=3)  # Sá»‘ chiá»u quan sÃ¡t (n_dim_obs) lÃ  3

# Dá»± Ä‘oÃ¡n vÃ  Æ°á»›c lÆ°á»£ng tráº¡ng thÃ¡i áº©n
filtered_state_means, _ = kf.filter(measurements)

# TrÃ­ch xuáº¥t dá»± Ä‘oÃ¡n cá»§a cá»™t '8', '9', '10'
predicted_values = filtered_state_means

# Trá»±c quan hÃ³a
plt.figure(figsize=(12, 18))

# Biá»ƒu Ä‘á»“ cho cá»™t '8'
plt.subplot(3, 1, 1)
plt.plot(df.index, df['8'], label='Actual (Column 8)', color='blue')
plt.plot(df.index, predicted_values[:, 0], label='Predicted (Column 8)', color='red')
plt.title('Actual vs Predicted Values for Column 8')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()

# Biá»ƒu Ä‘á»“ cho cá»™t '9'
plt.subplot(3, 1, 2)
plt.plot(df.index, df['9'], label='Actual (Column 9)', color='green')
plt.plot(df.index, predicted_values[:, 1], label='Predicted (Column 9)', color='orange')
plt.title('Actual vs Predicted Values for Column 9')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()

# Biá»ƒu Ä‘á»“ cho cá»™t '10'
plt.subplot(3, 1, 3)
plt.plot(df.index, df['10'], label='Actual (Column 10)', color='purple')
plt.plot(df.index, predicted_values[:, 2], label='Predicted (Column 10)', color='brown')
plt.title('Actual vs Predicted Values for Column 10')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()

# ÄÃ¡nh giÃ¡ mÃ´ hÃ¬nh
rmse_8 = np.sqrt(mean_squared_error(df['8'], predicted_values[:, 0]))
mae_8 = mean_absolute_error(df['8'], predicted_values[:, 0])

rmse_9 = np.sqrt(mean_squared_error(df['9'], predicted_values[:, 1]))
mae_9 = mean_absolute_error(df['9'], predicted_values[:, 1])

rmse_10 = np.sqrt(mean_squared_error(df['10'], predicted_values[:, 2]))
mae_10 = mean_absolute_error(df['10'], predicted_values[:, 2])

print("RMSE for Column 8:", rmse_8)
print("MAE for Column 8:", mae_8)

print("RMSE for Column 9:", rmse_9)
print("MAE for Column 9:", mae_9)

print("RMSE for Column 10:", rmse_10)
print("MAE for Column 10:", mae_10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Äá»c dá»¯ liá»‡u
df = pd.read_csv('bt.csv', index_col=0)

# Chia dá»¯ liá»‡u thÃ nh táº­p huáº¥n luyá»‡n vÃ  táº­p kiá»ƒm tra
train_data = df.iloc[:int(0.8*len(df))]
test_data = df.iloc[int(0.8*len(df)):]

# Khá»Ÿi táº¡o cÃ¡c biáº¿n Ä‘á»ƒ lÆ°u trá»¯ RMSE vÃ  MAE cho má»—i cá»™t
rmse_list = []
mae_list = []

# Huáº¥n luyá»‡n mÃ´ hÃ¬nh SARIMA vÃ  Ä‘Ã¡nh giÃ¡ cho tá»«ng cá»™t
for column in ['8', '9', '10']:
    # Huáº¥n luyá»‡n mÃ´ hÃ¬nh SARIMA cho cá»™t hiá»‡n táº¡i
    model = SARIMAX(train_data[column], exog=train_data.drop(columns=['8', '9', '10']), order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit(disp=False)
    
    # Dá»± Ä‘oÃ¡n vÃ  tÃ­nh toÃ¡n RMSE vÃ  MAE
    predicted_values = model_fit.forecast(len(test_data), exog=test_data.drop(columns=['8', '9', '10']))
    rmse = np.sqrt(mean_squared_error(test_data[column], predicted_values))
    mae = mean_absolute_error(test_data[column], predicted_values)
    
    # LÆ°u trá»¯ RMSE vÃ  MAE vÃ o danh sÃ¡ch
    rmse_list.append(rmse)
    mae_list.append(mae)
    
    # Trá»±c visual hÃ³a dá»± Ä‘oÃ¡n vÃ  thá»±c táº¿
    plt.figure(figsize=(20, 6))
    plt.plot(test_data.index, test_data[column], label='Actual ({})'.format(column), color='blue')
    plt.plot(test_data.index, predicted_values, label='Predicted ({})'.format(column), color='red')
    plt.title('Actual vs Predicted Values for {}'.format(column))
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

# In ra káº¿t quáº£ RMSE vÃ  MAE cho má»—i cá»™t
for i, column in enumerate(['8', '9', '10']):
    print("RMSE for Column {}: {:.4f}".format(column, rmse_list[i]))
    print("MAE for Column {}: {:.4f}".format(column, mae_list[i]))
```
